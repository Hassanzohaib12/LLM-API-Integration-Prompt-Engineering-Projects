```python
def ask_llm(prompt):
    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"Error: {e}"
```

```python
print("="*60)
print("ZERO-SHOT PROMPTING")
print("="*60)

prompt = """
Explain Artificial Intelligence in simple words for a beginner student.
"""

result = ask_llm(prompt)
print(result)
```
